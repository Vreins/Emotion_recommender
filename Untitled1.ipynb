```python
def book_recommender(book_title):

    book_index = books_df.loc[books_df['title']==book_title].index[0]
    book_index
    book_list = list(enumerate(similarity_mat[book_index]))
    book_list = sorted(book_list , key = lambda x:x[1] ,reverse=True)
    book_list = book_list[1:11]
    book_array = []

    for i in range(len(book_list)):
        idx = book_list[i][0]
        book_id= books_df["book_id"][idx]
        url = f"https://www.googleapis.com/books/v1/volumes/{book_id}"
        response = requests.get(url)
        data = response.json()
        info=data["volumeInfo"]
        book_array.append({
            "title": info.get("title"),
            "authors": ", ".join(info.get("authors", [])),
            "thumbnail": info.get("imageLinks", {}).get("thumbnail"),
            "preview": info.get("previewLink")
        })
    return book_array
```
